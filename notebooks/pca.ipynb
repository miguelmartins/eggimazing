```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples
N = 500

# Generate uncorrelated data in the rotated coordinate system
x_prime = np.random.randn(N)
y_prime = np.random.randn(N)
```

```python
# Rotation angle in radians (45 degrees)
theta = np.radians(45)

# Rotation matrix
cos_theta = np.cos(theta)
sin_theta = np.sin(theta)
R = np.array([
    [cos_theta, -sin_theta],
    [sin_theta,  cos_theta]
])

# Apply rotation
data_prime = np.vstack((x_prime, y_prime))  # Shape: (2, N)
data = R @ data_prime  # Shape: (2, N)

# Extract rotated data
x = data[0, :]
y = data[1, :]
```

```python
# Plot uncorrelated data in the rotated coordinate system
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x_prime, y_prime, alpha=0.5)
plt.xlabel("$x'$")
plt.ylabel("$y'$")
plt.title("Uncorrelated Data in Rotated Coordinate System")

# Plot correlated data in the canonical coordinate system
plt.subplot(1, 2, 2)
plt.scatter(x, y, alpha=0.5, color='orange')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Correlated Data in Canonical Coordinate System")

plt.tight_layout()
plt.show()
```

```python
# Inverse rotation (since R is orthogonal, inverse is transpose)
R_inv = R.T

# Reproject the data
data_reprojected = R_inv @ data

# Extract reprojected data
x_prime_recovered = data_reprojected[0, :]
y_prime_recovered = data_reprojected[1, :]
```

```python
# Compute correlation coefficients
corr_coeff_original = np.corrcoef(x_prime, y_prime)[0, 1]
corr_coeff_rotated = np.corrcoef(x, y)[0, 1]
corr_coeff_recovered = np.corrcoef(x_prime_recovered, y_prime_recovered)[0, 1]

print(f"Correlation coefficient in rotated coordinate system: {corr_coeff_original:.3f}")
print(f"Correlation coefficient in canonical coordinate system: {corr_coeff_rotated:.3f}")
print(f"Correlation coefficient after reprojecting: {corr_coeff_recovered:.3f}")
```

```python
# Plot reprojected data
plt.figure()
plt.scatter(x_prime_recovered, y_prime_recovered, alpha=0.5)
plt.xlabel("$x'$ (recovered)")
plt.ylabel("$y'$ (recovered)")
plt.title("Reprojected Data (Uncorrelated)")
plt.show()
```
